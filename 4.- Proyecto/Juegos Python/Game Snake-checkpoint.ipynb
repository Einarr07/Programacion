```python
import turtle 
import time
import random
posponer = 0.1

#Marcador 
score = 0
high_score = 0

#Ventana
ventana = turtle.Screen()
ventana.title("Juego de la culebrita")
ventana.bgcolor("light green")
ventana.setup(width = 600, height = 600)
ventana.tracer(0)

#Cabeza Snake
cabeza = turtle.Turtle()
cabeza.speed(0) #Se presente la cabeza en pantalla
cabeza.shape("square") #Forma de la cabeza
cabeza.color("green") #Color cabeza
cabeza.penup() #Borrar rastro
cabeza.goto(0,0) #Iniciar desde el centro de la pantalla
cabeza.direction = "stop" #Esperar que le de una direccion

#Comida
comida = turtle.Turtle()
comida.speed(0) #Se presente la comida en pantalla
comida.shape("circle") #Forma de la comida
comida.color("red") #Color comida
comida.penup() #Borrar rastro
comida.goto(0,100) #Iniciar desde el centro de la pantalla

#Segmentos / Cuerpo de la serpiente
segmentos = []

#Texto en pantalla
texto = turtle.Turtle()
texto.speed(0) #Se genera el texto inmediatamente
texto.color("white")
texto.penup() #Evitamos la pluma que pinta el texto
texto.hideturtle() #Escondemos la pluma
texto.goto(0,260) #Lugar donde aparece el texto
texto.write("Score: 0     High Score: 0", align = "center", font =("Courier",24,"normal"))

#Funciones movimiento
def arriba():
    cabeza.direction = "up"
def abajo():
    cabeza.direction = "down"
def izquierda():
    cabeza.direction = "left"
def derecha():
    cabeza.direction = "right"
#
def mov():
    if cabeza.direction == "up": #Hacia arriba
        y = cabeza.ycor() #cordenada cabeza
        cabeza.sety(y + 20)
    
    if cabeza.direction == "down": #Hacia abajo
        y = cabeza.ycor() #cordenada cabeza
        cabeza.sety(y - 20)
    
    if cabeza.direction == "left": #Hacia la izquierda
        x = cabeza.xcor() #cordenada cabeza
        cabeza.setx(x - 20)
    
    if cabeza.direction == "right": #Hacia la derecha
        x = cabeza.xcor() #cordenada cabeza
        cabeza.setx(x + 20)
#Teclado
ventana.listen()
ventana.onkeypress(arriba, "Up")
ventana.onkeypress(abajo, "Down")
ventana.onkeypress(izquierda, "Left")
ventana.onkeypress(derecha, "Right")

while True:
    ventana.update()
    
    #Colisiones con los bordes
    if cabeza.xcor() > 280 or cabeza.xcor() < -280 or cabeza.ycor() > 280 or cabeza.ycor() < -280:
        time.sleep(1)
        cabeza.goto(0,0)
        cabeza.direction = "stop"
        
        #Esconder los segmentos / cuerpo de la snake
        for segmento in segmentos: 
            segmento.goto(1000,1000)
        
        #Limpiar lista de segmentos
        segmentos.clear()
        
        #Resetear marcador
        score = 0
        texto.clear()
        texto.write("Score: {}     High Score: {}".format(score, high_score), 
                    align = "center", font =("Courier",24,"normal"))
    #Colision comida
    if cabeza.distance(comida) < 20: #Colision 
        x = random.randint(-280,280) #Lugar de la comida
        y = random.randint(-280,280)
        comida.goto(x,y)
        Nuevo_Segmento = turtle.Turtle()
        Nuevo_Segmento.speed(0) #Se presente la Nuevo_Segmento en pantalla
        Nuevo_Segmento.shape("square") #Forma de la Nuevo_Segmento
        Nuevo_Segmento.color("yellow") #Color Nuevo_Segmento
        Nuevo_Segmento.penup() #Borrar rastro
        segmentos.append(Nuevo_Segmento)
        
        #Aumentar marcador
        score += 10
        if score > high_score:
            high_score = score
        
        texto.clear()
        texto.write("Score: {}     High Score: {}".format(score, high_score), 
                    align = "center", font =("Courier",24,"normal"))
    #Mover el cuerpo de la serpiente
    totalSeg = len(segmentos)
    for index in range (totalSeg -1,0,-1):
        x = segmentos[index -1].xcor()
        y = segmentos[index -1].ycor()
        segmentos[index].goto(x,y)
    if totalSeg > 0:
        x = cabeza.xcor()
        y = cabeza.ycor()
        segmentos[0].goto(x,y)
    
    mov()
    
    #Colisiones con el cuerpo
    for segmento in segmentos:
        if segmento.distance(cabeza) < 20:
            time.sleep(1)
            cabeza.goto(0,0)
            cabeza.direction = "stop"
        
            #Esconder los segmentos / cuerpo de la snake
            for segmento in segmentos: 
                segmento.goto(1000,1000)
        
            #Limpiar lista de segmentos
            segmentos.clear()
            #Resetear marcador con el cuerpo
            score = 0
            texto.clear()
            texto.write("Score: {}     High Score: {}".format(score, high_score), 
                        align = "center", font =("Courier",24,"normal"))
        
    time.sleep(posponer) #Disminuir el timpo de programa    
```
